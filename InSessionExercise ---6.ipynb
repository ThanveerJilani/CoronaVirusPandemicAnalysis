```python
# Essential packages.
import pandas as pd
```

```python
# Load the dataset.
df = pd.read_csv('covid_19_india.csv')
```

```python
# Displaying first five rows.
df.head()
```

```python
# Removing unnecessary columns.
df = df.drop(columns = ["Sno" , "Date" , "ConfirmedIndianNational" , "ConfirmedForeignNational" , "Confirmed" , "Time"])
df
```

```python
# Renaming the column.
df = df = df.rename(columns={"State/UnionTerritory": "States"})
df
```

```python
# Selecting only "Kerala" in the column. 
filt = df["States"] == "Kerala"
df = df[filt]
df
```

```python
# Correlation of all columns in the dataframe.
correlation_df = df.corr()
```

```python
# Displaying onto the console.
correlation_df
```

```python
# Correlation of two columns --- (1)
column_1 = df["Cured"]
column_2 = df["Deaths"]
correlation = column_1.corr(column_2)
```
