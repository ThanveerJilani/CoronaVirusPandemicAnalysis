```python
# We will remove Sno , Time , ConfirmedIndianNational , ConfirmedForeignNational columns with drop() function
df = df.drop(columns = [ "Sno" , "Time" , "ConfirmedIndianNational" , "ConfirmedForeignNational"])
df
```

```python
# Now , We'll rename State/UnionTerritory as it is very long.
df = df = df.rename(columns={"State/UnionTerritory": "States"})
df
```

```python
# We need only Delhi state. We are going to fiter the data.

filt = df["States"] == "Delhi"
df = df[filt]
df
```

```python
# We need data from 1st of June to 30th june so, We are going to use list slicing.
# There are many methods to do so, but I'm focussing on simple methods.
df = df[91:-2]
df
```

```python
# Now , We have some good data.
# We'll remove the state as it is not required now using drop() function.
df = df.drop(columns = "States")
df
```
