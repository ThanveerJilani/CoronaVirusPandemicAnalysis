```python
# Importing all essential packages.
from tkinter import *
import tkinter
```

```python
# Declaration
root = Tk()
root.title("Pandemic Analytics Search Report")
root.geometry("700x800")

# Declaring a function that holds the total population , total recovered and deaths.
def cal_func():

    total_population = entry2.get()
    int_total_population = int(total_population)

    infected = entry3.get()
    int_infected = int(infected)

    death = entry4.get()
    int_death = int(death)

    ifr = (int_death / int_infected) * 100
    cmr = (int_death / int_total_population)
    
    
    
    root = Tk()
    root.title("Result")

    label1 = Label(root, font=("helvetica", 22, "bold"), text="Result" , bg = "grey" , fg = "black")
    label1.grid(row=7, columnspan=1)

    label2 = Label(root, font=("times", 14, "bold"), text="IFR : ", bg = "grey" , fg = "black")
    label2.grid(row=9, column=1)
    label3 = Label(root, text=ifr)
    label3.grid(row=9, column=2)

    label4 = Label(root, font=("times", 14, "bold"), text="CMR : ", bg = "grey" , fg = "black")
    label4.grid(row=10, column=1)
    label5 = Label(root, text=cmr)
    label5.grid(row=10, column=2)

# Listing out all the states in alphabetical order.    
options = ['Andaman and Nicobar', 'Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar', 'Chandigarh', 'Chhattisgarh', 'Dadra and Nagar Haveli', 'Delhi', 'Goa', 'Gujarat', 'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir', 'Jharkhand', 'Karnataka', 'Kerala', 'Ladakh', 'Lakshadweep', 'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Odisha', 'Puducherry', 'Punjab', 'Rajasthan', 'Sikkim', 'Tamil Nadu', 'Telengana', 'Tripura', 'Uttar Pradesh', 'Uttarakhand', 'West Bengal']
select = StringVar() 
select.set(options[0])

# Setting font and colour.
label1 = Label(root,font=("helvetica", 22, "bold"), text="Pandemic Analytics Report",bg = "grey" , fg = "black")
label1.grid(row=1,columnspan = 1)

label2 = Label(root, text="Enter the State:",bg = "grey" , fg = "black")
label2.grid(row=2)
drop = OptionMenu(root,select,*options)
drop.grid(row=2,column=2)

label3 = Label(root, text="Total population:" , bg = "grey" , fg = "black")
label3.grid(row=3)
entry2 = Entry(root)
entry2.grid(row=3, column=2)

label4 = Label(root, text="Enter Infected:" , bg = "grey" , fg = "black")
label4.grid(row=4)
entry3 = Entry(root)
entry3.grid(row=4, column=2)

label5 = Label(root, text="Enter deaths:" , bg = "grey" , fg = "black")
label5.grid(row=5)
entry4 = Entry(root)
entry4.grid(row=5, column=2)


# Button for calculating the IFR and CMR.
button1 = Button(root, text="Calculate IFR And CMR:",command=cal_func , bg = "grey" , fg = "black")
button1.grid(column=2)
# Displaying the GUI on to the console.
root.mainloop()
```
